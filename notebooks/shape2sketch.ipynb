```python
from lib.utils import load_config
from lib.data.preprocess_sketch import (
    render_shapenet,
    image_grid,
    image_to_sketch,
    obj_path,
    cartesian_elev_azim,
)
from lib.data.metainfo import MetaInfo

config = load_config("config")
metainfo = MetaInfo(config.data.data_dir) 
obj1 = metainfo.obj_path(metainfo.obj_ids[0])
# obj2 = obj_path(config.obj_ids[1], config=config)

elev, azim = cartesian_elev_azim(elev=config.data.elev, azim=config.data.azim)
images = render_shapenet(obj1, color=0.4, elev=elev, azim=azim)
image_grid(images, rows=5, cols=10)
```
