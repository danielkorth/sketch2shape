```python
masks = np.load("/home/korth/sketch2shape/temp/opt_render/mask_prior_19_acc.npy")
std = np.std(masks, axis=0)
mean = np.mean(masks, axis=0)
ci = 1.96 * std / masks.shape[0]
upper = mean + ci
lower = mean - ci


fig, ax = plt.subplots()
x = list(range(100))
ax.plot(x, mean)
ax.set_yscale('log')
ax.fill_between(
    x, lower, upper, color='b', alpha=.15)
ax.set_ylim(ymin=0)
ax.set_title('mask.sum() at every render step')
fig.autofmt_xdate(rotation=45)
```
